```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
payoff = lambda mu, alpha, freq_mu, freq_alpha: np.power((mu+alpha),(freq_mu-freq_alpha))
```

```python
def mean_median_std_plot(combinaison, res):
    all_values = list(res.values())
    mean = np.mean(all_values)
    median = np.median(all_values)
    std = np.std(all_values)
    max_val = np.max(all_values)
    min_val = np.min(all_values)
    fig, ax = plt.subplots(figsize=(15, 15))
    ax.bar([str(x) for x in combinaison.keys()], combinaison.values(), color ='maroon',
            width = 0.4)
    fig.autofmt_xdate(rotation=45)
    return {'mean':mean, 'median':median, 'std':std, 'max':max_val, 'min':min_val}
```

```python
combinaison = Counter()
res = {}
for a in range(6):
    for b in range(6):
        for c in range(6):
            for d in range(6):
                for e in range(6):
                    for f in range(6):
                        l = [a+1, b+1, c+1, d+1, e+1, f+1]
                        dic = dict(Counter(l))
                        new_dic = {1: [], 2:[], 3:[], 4:[], 5:[], 6:[]}
                        for key, value in dic.items():
                            new_dic[value].append(key)
                        for freq in range(6, 0, -1):
                            if len(new_dic[freq])>0:
                                max_freq = freq
                                max_freq_value = max(new_dic[freq])
                                break
                        for freq in range(1, 7):
                            if len(new_dic[freq])>0:
                                min_freq = freq
                                min_freq_value = min(new_dic[freq])
                                break
                        if min_freq_value == max_freq_value:
                            min_freq_value = 0
                            min_freq = 0
                        val = payoff(max_freq_value, min_freq_value, max_freq, min_freq)
                        res[str(a+1)+str(b+1)+str(c+1)+str(d+1)+str(e+1)+str(f+1)] = val
                        combinaison[val] +=1
combinaison = dict(sorted(combinaison.items(), key=lambda item: item[1], reverse=True))
```

```python
mean_median_std_plot(combinaison, res)
```
